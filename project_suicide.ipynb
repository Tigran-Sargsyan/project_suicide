# Նկարագրություն

##  Happiness report dataset
> Առաջատար փորձագետները տարբեր ոլորտներում (տնտեսագիտություն, հոգեբանություն, հարցումների վերլուծություն, ազգային վիճակագրություն, առողջապահություն, հանրային քաղաքականություն և այլն), նկարագրում են, թե ինչպես կարող են բարեկեցության չափումները արդյունավետորեն օգտագործվել ազգերի առաջընթացը գնահատելու համար: Զեկույցները ներկայացնում են երջանկության վիճակը 2015-2019թթ աշխարհում և ցույց են տալիս, թե ինչպես է երջանկության նոր գիտությունը բացատրում երջանկության անձնական և ազգային տատանումները:


*   Country - Երկիր
*   Happiness Score - Երջանկության ցուցանիշ
*   Economy (GDP per Capita) - Տնտեսություն (1 շնչի հաշվով ՀՆԱ)
*   Social support - Սոցիալական աջակցություն
*   Healthy life expectancy - Առողջ կյանքի տևողությունը
*   Freedom to make life choices - Կյանքի ընտրություն կատարելու ազատություն
*   Generosity - Առատաձեռնություն
*   Perceptions of corruption - Կոռուպցիայի ընկալումը
*   Country Code - Երկրի կոդը

## HDI (Human development index)
> Մարդկային զարգացման ինդեքսը կյանքի տեւողության, կրթության և մեկ շնչին ընկնող եկամտի ցուցանիշների վիճակագրական կոմպոզիտային ինդեքս է, որն օգտագործվում է երկրները մարդկային զարգացման չորս մակարդակներում դասակարգելու համար:

* HDI (Human development index) - Մարդկային զարգացման ցուցանիշ

## Suicides dataset

```python
import pandas as pd

url = 'https://drive.google.com/file/d/16dTcedcMSIom1nvcZp2dUUG_mNcSD9Ek/view?usp=sharing'
file_id = url.split('/')[-2]
export_url = f'https://drive.google.com/uc?export=download&id={file_id}'
suicide = pd.read_csv(export_url)
```

##Human development index

```python
import pandas as pd

url = 'https://drive.google.com/file/d/1Oju1mVVkoOLNDpSRSpoah5jhALehAaGv/view?usp=sharing'
file_id = url.split('/')[-2]
export_url = f'https://drive.google.com/uc?export=download&id={file_id}'
HDI = pd.read_csv(export_url, skiprows=1)
```

```python
HDI.drop(columns=['HDI Rank (2018)'], inplace=True)
HDI.drop([i for i in range(189,212)], inplace=True)
HDI[[str(i) for i in range(1990,2019)]] = (HDI.loc[:,"1990":]).apply(pd.to_numeric, errors="coerce", axis=1)
HDI.shape
```

```python
hdi = HDI.merge(regions[["Country","Country Code"]], on="Country", how ="left")
hdi[["Country","Country Code"]][hdi[["Country","Country Code"]].isnull().any(1)]
hdi.at[[20,21,39,44,45,56,70,74,79,90,93,113,126,141,165,167,184,185],["Country Code"]] = ["BOL","BIH","COD","CZE","CIV","SWZ","GNB","HKG","IRN",
                                                                                           "KOR","LAO","MDA","MKD","RUS","SYR","TZA","VEN","VNM"]  
```

## Finding correlations

Գտնում ենք տվյալների կոռելացիոն գործակիցները՝ նրանց միջև առկա կապերն ուսումնասիրելու համար։

```python
from itertools import combinations

cols = data_new.columns.values[4:]
pairs = list(combinations(cols,2))
cor_lst = []

def cor(X,Y):
    my_cor = data_new[X].corr(data_new[Y]).round(5)
    my_tuple = X,Y,my_cor
    cor_lst.append(my_tuple)

for pair in pairs:
    X,Y = pair[0],pair[1]
    cor(X,Y)
```

```python
cor_lst.sort(key=lambda x:x[2], reverse = True)
```

### Filtering correlations

```python
very_strong_pos = list(filter(lambda x: 0.7 < x[2] <= 1,cor_lst))
strong_pos = list(filter(lambda x: 0.5 < x[2] <= 0.7,cor_lst))
moderate_pos = list(filter(lambda x: 0.3 < x[2] <= 0.5,cor_lst))
weak_pos = list(filter(lambda x: 0 < x[2] <= 0.3,cor_lst))
```

```python
very_strong_neg = list(filter(lambda x: -1<=x[2]<-0.7,cor_lst))
strong_neg = list(filter(lambda x: -0.7<=x[2]<-0.5,cor_lst))
moderate_neg = list(filter(lambda x: -0.5<=x[2]<-0.3,cor_lst))
weak_neg = list(filter(lambda x: -0.3<=x[2]<0,cor_lst))
```

```python
data_new_corr_prs = data_new.loc[:,'Happiness Score':]
data_new_corr_prs = data_new_corr_prs.corr(method='pearson')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,8),linewidth=10,edgecolor="#04253a" )
sns.heatmap(data_new_corr_prs, annot=True,cmap='Blues')
plt.show()
```

```python
data_new_corr_spr = data_new.loc[:,'Happiness Score':]
data_new_corr_spr = data_new_corr_spr.corr(method='spearman')

plt.figure(figsize=(10,8),linewidth=10,edgecolor="#04253a" )
sns.heatmap(data_new_corr_spr, annot=True,cmap='YlOrRd')
plt.show()
```

```python
data_new_corr_knd = data_new.loc[:,'Happiness Score':]
data_new_corr_knd = data_new_corr_spr.corr(method='kendall')

plt.figure(figsize=(10,8),linewidth=10,edgecolor="#04253a" )
sns.heatmap(data_new_corr_knd, annot=True,cmap='Greens')
plt.show()
```
